### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  
2. Remova as variáveis não significantes.  
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
tips = load_dataset(name='tips')

tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']

tips
````

````python
# Criando a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de tip, explicado por sex, smoker, diner e net_bill.

y, X = patsy.dmatrices(formula_like='tip ~ sex + smoker + time + net_bill', 
                       data=tips)
X
````

````python
y
````

````python
# Removendo as variáveis não significantes.
y, X = patsy.dmatrices('tip_pct ~ net_bill', tips)

mod = sm.OLS(y, X).fit()
mod.summary()
````

````python
# Observando o gráfico de resíduos em função de net_bill

sns.scatterplot(x = 'net_bill', y = mod.resid, data = tips)
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
````

````python
# Testando transformar net_bill no log e um polinômio. 

y, X = patsy.dmatrices(formula_like='tip ~ np.log(net_bill)', data=tips)

mod_log = sm.OLS(y, X).fit()
mod_log.summary()
````

O primeiro modelo (mod) apresentou melhor resultado com um coefiente de determinação maior que o segundo modelo (mod_log).

### II. Modelo no valor do percentual da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de ```tip```, explicado por ```sex, smoker, diner e net_bill```.
2. Remova as variáveis não significantes.
3. Observe o gráfico de resíduos em função de ```net_bill```
4. Teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
5. Do modelo final deste item, calcule o $R^2$ na escala de ```tip``` (sem o log). Compare com o modelo do item 1. Qual tem melhor coeficiente de determinação?

````python
# Criando a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de tip, explicado por sex, smoker, diner e net_bill.

y, X = patsy.dmatrices('np.log(tip) ~ sex + smoker + time + net_bill', data=tips)

mod1 = sm.OLS(y, X).fit()
mod1.summary()
````

````python
# Removendo as variáveis não significantes.

y, X = patsy.dmatrices(formula_like='np.log(tip) ~ net_bill', data=tips)

mod1 = sm.OLS(y, X).fit()
mod1.summary()
````

````python
# Observando o gráfico de resíduos em função de net_bill

sns.scatterplot(x='net_bill', y=mod1.resid, data=tips)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
# Testando transformar net_bill no log e um polinômio. 

# Primeiramente em log
y, X = patsy.dmatrices(formula_like='np.log(tip) ~ np.log(net_bill)', data=tips)

mod1_log = sm.OLS(y, X).fit()
mod1_log.summary()
````

````python
# Segundo em polinômio

y, X = patsy.dmatrices(formula_like='np.log(tip) ~ net_bill + np.power(net_bill,2)', data=tips)

mod1_pow = sm.OLS(y, X).fit()
mod1_pow.summary()
````

Pode-se dizer que após a transformação, o que mudou para polinômio apresentou maior coeficiente de determinação, por conseguinte, melhor resultado. 

````python
# Do modelo final deste item, calcule o  𝑅2  na escala de tip (sem o log).
print( 'Comparando os R2:')

tips['pred_tip_log'] = np.exp(mod1_log.fittedvalues)
print('R2 do modelo em log:',tips[['pred_tip_log', 'tip']].corr().iloc[0,1]**2)

tips['pred_tip_pow'] = np.exp(mod1_pow.fittedvalues)
print('R2 do modelo em polinômio :',tips[['pred_tip_pow', 'tip']].corr().iloc[0,1]**2)
````

Mais uma vez o modelo transformado em polinômio apresentou R2 superior ao transformado em log.
