# Regressão 01 - tarefa 01

### 1. Explorando a base de gorjetas

a. Na aula fizemos uma regressão de ```tip``` explicada por ```net_bill```. Carregue os pacotes e a base, crie as variáveis necessárias como na aula, e reproduza esta regressão.

b. Reproduza também o gráfico de dispersão dessas duas variáveis.

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.formula.api as smf

from sklearn.metrics import r2_score

%matplotlib inline
````

````python
tips = sns.load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
# Reproduzindo o gráfico de dispersão dessas duas variáveis.

sns.regplot(y = 'tip', x = 'net_bill', data = tips[tips['tip_pct']<.5])

# Gerando dados para a linha dos 10%
x = np.linspace(0,tips['net_bill'].max(),100)
y = .1*x

# Plotando a linha dos 10%
plt.plot(x, y, '-r', label='linha dos 10%')

# Opções estéticas para explicitar o (0,0) no gráfico
plt.xlim(left = 0)
plt.ylim(bottom = 0)
````

### 2. Mudança de perspectiva

Agora ajuste um modelo de regressão de ```tip_pct``` por ```net_bill```. Construa o gráfico de dispersão correspondente.

````python
sns.regplot(y='tip_pct',x='net_bill', data=tips[tips['tip_pct']<.5])
x = np.linspace(0,tips['net_bill'].max(),100)
y = .1*x
plt.plot(x, y, '-r', label='linha dos 10%')
plt.xlim(left = 0)
plt.ylim(bottom = 0)
````

### 3. Comente

Compare os dois gráficos e comente (de forma objetiva e suscinta) a relação entre eles e a diferença do ajuste. Podemos comparar diretamente os $R^2$s?

````python
reg = smf.ols('tip ~ net_bill', data = tips).fit()
reg.summary()
````

````python
reg = smf.ols('tip_pct ~ net_bill', data = tips).fit()
reg.summary()
````

Em relação aos gráficos pode-se concluir:
. ambos apresentam baixo coeficiente de determinação (R2)
. o primeiro gráfico apresenta uma correlação positiva, enquanto o segundo gráfico apresenta uma correlação inversa.

### 4. Ajuste do modelo em outra escala

Com o modelo no item 3, obtenha a preditiva do valor da gorjeta:
  1. obtenha o valor predito ```pred``` para ```tip_pct``` (dica, veja o atributo ```reg.fittedvalues```)
  2. obtenha o valor predito para ```tip``` como ```tip_pct * net_bill```
  3. calcule o $R^2$ do modelo para ```tip```
  4. compare esse $R^2$ com o do item 1

````python
# Obtendo o valor predito pred para tip_pct

reg_tip_pct = smf.ols(formula='tip_pct ~ net_bill', data=tips).fit()

reg_tip_pct.fittedvalues
````

````python
# Obtendo o valor predito para tip como tip_pct * net_bill
reg_tip_pct.fittedvalues * tips.net_bill
````

````python
# Calculando o  𝑅2 do modelo para tip

r2 = r2_score(y_true=tips['tip'], 
              y_pred=reg_tip_pct.fittedvalues * tips.net_bill)

print('R2 do modelo para tip:', r2)
````

````python
# Comparando esse  𝑅2 com o do item 1

print('𝑅2 dos valores preditos pred para tip_pct:', 
      round(r2 * 100, 2), '%')

print('𝑅2 do modelo `reg_tip_pct`:', 
      round(reg_tip_pct.rsquared * 100, 2), '%')
````
