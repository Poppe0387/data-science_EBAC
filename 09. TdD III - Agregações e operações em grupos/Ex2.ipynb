### 1. Crie 2 faixas de Latitude do município (munResLat) sendo uma acima e outra abaixo de -10.5 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

```python
minLat =sinasc_raw.munResLat.min()
```

```python
maxLat =sinasc_raw.munResLat.max()
```

```python
faixasLat = pd.cut(sinasc_raw['munResLat'], [minLat, -10.5,maxLat])
```

```python
pd.pivot_table(sinasc_raw, values=['QTDFILVIVO', 'QTDFILMORT'], index = faixasLat, aggfunc= ['sum','mean','min','max','median','std',np.var], dropna =False)
```

```python
minArea =sinasc_raw.munResArea.min()
```

```python
maxArea =sinasc_raw.munResArea.max()
```

```python
faixasArea = pd.cut(sinasc_raw['munResArea'], [minArea, 3000,maxArea])
```

```python
pd.pivot_table(sinasc_raw, values=['IDADEMAE', 'IDADEPAI'], index = faixasArea, aggfunc= ['sum','mean','min','max','median','std',np.var], dropna =False)
```
