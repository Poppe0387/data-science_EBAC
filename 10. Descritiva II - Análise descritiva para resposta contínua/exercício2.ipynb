````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

np.random.seed(2360873)
N = 100
````

````python
df = pd.read_csv('profissao.csv')
df
````

````python
sns.pointplot(x="posse_de_veiculo", y="renda", data=df, dodge=True)
````

````python
sns.pointplot(x="posse_de_imovel", y="renda", data=df, dodge=True)
````

````python
sns.pointplot(x="posse_de_veiculo", y="renda",hue='posse_de_imovel', data=df, dodge=True)
````

````python
sns.pointplot(x="posse_de_imovel", y="renda",hue='posse_de_veiculo', data=df, dodge=True)
````

A partir da análise acima podemos dizer que as duas variáveis (posse do veiculo e posse  do imóvel) discriminam a renda de forma significativa.
No entanto, a importância predominante na previsão de renda é atribuída à variável de posse de veículo, devido à significativa disparidade no intervalo de confiança entre indivíduos que possuem veículo e aqueles que não possuem.
