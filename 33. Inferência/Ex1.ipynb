## 1) Diferena de peso entre machos e fmeas

Na base de pinguins, podemos ver que h diferena de peso entre machos e fmeas. Mas ser que  uma concluso extrapolvel a toda a populao de pinguins? Assumindo que essa amostra  aleatria e representativa da populao de pinguins, realize um teste T com 5% de significncia para verificar se podemos induzir esta concluso para a populao toda.

```python
desvpad = pg.groupby('sex')['body_mass_g'].var().mean()**.5
Mdia_A = pg.groupby('sex')['body_mass_g'].mean()['Male']
Mdia_B = pg.groupby('sex')['body_mass_g'].mean()['Female']

diferena = Mdia_A - Mdia_B # Numerador da frmula
erro_padrao = desvpad*((2/10)**.5) # varincia amostral*raiz(2/n)
t = diferena/erro_padrao # Estatstica do teste
RC = st.t.ppf(.95, 18) # Regio crtica: sob h0, t est na RC 5% das vezes
p_value = 1-st.t.cdf(t, 18) # p-value: a menor significncia para a qual rejeitamos o teste

print(f'diferena: {diferena:.2f}')
print(f'desvio padro: {desvpad:.2f}')
print(f'erro padro: {erro_padrao:.2f}')
print(f'estatstica do teste: {t:.2f}')
print(f'regio crtica: {RC:.2f} ou mais')
print(f'p-value: {p_value:.5f} ou mais')
```

Conclui-se que o resultado  estatisticamente relevante porque o p-value (0.02529)  menor que o nvel de significncia usual de 0.05. Isso significa que h menos de 2.529% de probabilidade de que a diferena observada na massa corporal entre machos e fmeas da espcie Gentoo seja devido ao acaso.

```python
# Espcie Adelie
desvpad = pg.loc[pg['species'] == 'Adelie'].groupby('sex')['body_mass_g'].var().mean()**.5
Mdia_A = pg.loc[pg['species'] == 'Adelie'].groupby('sex')['body_mass_g'].mean()['Male']
Mdia_B = pg.loc[pg['species'] == 'Adelie'].groupby('sex')['body_mass_g'].mean()['Female']

diferena = Mdia_A - Mdia_B # Numerador da frmula
erro_padrao = desvpad*((2/10)**.5) # varincia amostral*raiz(2/n)
t = diferena/erro_padrao # Estatstica do teste
RC = st.t.ppf(.95, 18) # Regio crtica: sob h0, t est na RC 5% das vezes
p_value = 1-st.t.cdf(t, 18) # p-value: a menor significncia para a qual rejeitamos o teste

print(f'diferena: {diferena:.2f}')
print(f'desvio padro: {desvpad:.2f}')
print(f'erro padro: {erro_padrao:.2f}')
print(f'estatstica do teste: {t:.2f}')
print(f'regio crtica: {RC:.2f} ou mais')
print(f'p-value: {p_value:.5f} ou mais')

p_value_adelie = p_value
```

```python
def calc_stats(species_name):
    desvpad = pg.loc[pg['species'] == species_name].groupby('sex')['body_mass_g'].var().mean()**.5
    Mdia_A = pg.loc[pg['species'] == species_name].groupby('sex')['body_mass_g'].mean()['Male']
    Mdia_B = pg.loc[pg['species'] == species_name].groupby('sex')['body_mass_g'].mean()['Female']

    diferena = Mdia_A - Mdia_B
    erro_padrao = desvpad*((2/10)**.5)
    t = diferena / erro_padrao
    p_value = 1 - st.t.cdf(t, 18)
    
    return diferena, p_value

# Espcies para anlise
species_list = ['Gentoo', 'Chinstrap', 'Adelie']

# Coletando os resultados
differences = []
p_values = []

for species in species_list:
    diferena, p_value = calc_stats(species)
    differences.append(diferena)
    p_values.append(p_value)

# Grfico de barras
fig, ax = plt.subplots(2, 1, figsize=(8, 10))

# Grfico de diferenas de mdias
ax[0].bar(species_list, differences, color='blue')
ax[0].set_title('Diferena de Mdia de Massa Corporal entre Machos e Fmeas')
ax[0].set_ylabel('Diferena de Massa Corporal (g)')

# Grfico de p-values
ax[1].bar(species_list, p_values, color='green')
ax[1].set_title('p-value para a Diferena de Massa Corporal')
ax[1].set_ylabel('p-value')
ax[1].axhline(0.05, color='red', linestyle='--')  # Linha de significncia 0.05

plt.tight_layout()
plt.show()
```
