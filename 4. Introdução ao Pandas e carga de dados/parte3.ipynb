3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
def consultar_solicitacoes(uf,tipo_interessado, DataFrame):
    DataFrame = pd.DataFrame(DataFrame)
    
    # Filtrar os dados de acordo com o UF da consulta e o tipo de interessado
    df = DataFrame[(DataFrame["UF"] == uf) & (DataFrame["tipo_interessado"] == tipo_interessado)]
    
    return df
````

````python
def df_bahia (uf, DataFrame):
    df_bahia = DataFrame[DataFrame["uf"]=="BA"]
    return df_bahia
````

````python
def total (uf, DataFrame):
    total = DataFrame[DataFrame["uf"] == uf] ["Status"].eq("Arquivado por decurso de prazo").sum()
    return total
````

````python
def solicitacoes_uf(uf, DataFrame):
    total = DataFrame[DataFrame["UF"] == uf] ["Status"].eq("Arquivado por decurso de prazo").sum()
    return solicitacoes_uf

def municipio_com_mais_deferidas(uf, DataFrame):
    df_uf = DataFrame(DataFrame["UF"] == uf)
    municipio_mais_deferido = df_uf[df_uf["Status"] == "Deferido"]["Município"].value_counts().idxmax()
    return municipio_mais_deferido

def csv_bahia(DataFrame):
    df_bahia = DataFrame[DataFrame["UF"]=="BA"]
    df_bahia.to_csv("solicitacoes_bahia.csv", index=False)


````
